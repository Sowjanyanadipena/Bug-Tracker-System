```python
import pandas as pd

# --- Initialize Bug List (In-memory storage) ---
bugs = []

# --- Function to Display Bugs ---
def display_bugs():
    if not bugs:
        print("\nNo bugs found.")
        return
    df = pd.DataFrame(bugs)
    print("\n--- Bug List ---")
    print(df.to_string(index=False))

# --- Function to Add Bug ---
def add_bug():
    bug_id = len(bugs) + 1
    title = input("Enter bug title: ")
    description = input("Enter bug description: ")
    severity = input("Enter severity (Low/Medium/High): ")
    status = "Open"
    assigned_to = input("Enter assigned person: ")
    bug = {
        "Bug ID": bug_id,
        "Title": title,
        "Description": description,
        "Severity": severity,
        "Status": status,
        "Assigned To": assigned_to
    }
    bugs.append(bug)
    print("\nBug added successfully!")

# --- Function to Update Bug Status ---
def update_bug():
    bug_id = int(input("Enter Bug ID to update: "))
    for bug in bugs:
        if bug["Bug ID"] == bug_id:
            print(f"Current Status: {bug['Status']}")
            new_status = input("Enter new status (Open/In Progress/Resolved/Closed): ")
            bug["Status"] = new_status
            print("\nBug status updated successfully!")
            return
    print("Bug not found.")

# --- Function to Delete Bug ---
def delete_bug():
    bug_id = int(input("Enter Bug ID to delete: "))
    for bug in bugs:
        if bug["Bug ID"] == bug_id:
            bugs.remove(bug)
            print("\nBug deleted successfully!")
            return
    print("Bug not found.")

# --- Main Menu ---
while True:
    print("\n=== Bug Tracker System ===")
    print("1. Add Bug")
    print("2. View Bugs")
    print("3. Update Bug Status")
    print("4. Delete Bug")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_bug()
    elif choice == "2":
        display_bugs()
    elif choice == "3":
        update_bug()
    elif choice == "4":
        delete_bug()
    elif choice == "5":
        print("Exiting Bug Tracker. Goodbye!")
        break
    else:
        print("Invalid choice! Please try again.")
```
